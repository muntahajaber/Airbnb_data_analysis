```python
## scroll down until the end of the page

def scroll_page(driver):
    lenOfPage = driver.execute_script("window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;")
    match=False
    while(match==False):
        lastCount = lenOfPage
        time.sleep(3)
        lenOfPage = driver.execute_script("window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;")
        if lastCount==lenOfPage:
            match=True
```

```python
## find all links for all houses 

def find_links_of_a_page(soup):
    links_a = soup.find_all('a', class_='_1jhvjuo')
    links = []
    for l in links_a:
        links.append(root + l.get('href'))
    return links
```

```python
## find house type and links 

def get_houses_types_and_urls(soup):
    house_types = set(soup.find_all('div', {'class' : "_j1kt73"}))
    house_types = set([i.text for i in house_types if i.text != 'More'])
    links = []
    for h in house_types:
        links.append(url+tag_ctg+tag_list[h])
    return house_types, links
```

```python
## convert the price to an integer

def price_to_int(moneystring):
    moneystring = moneystring.replace('SR', '').replace(',', '')
    return int(moneystring)
```

```python
## convert the str to an date

def str_to_date(string):
    return datetime.datetime.strptime(string, '%m/%d/%Y').date()
```

```python
def get_rating(soup):
    rating = soup.find('span', class_='_1ne5r4rt')
    if rating != None:
        return float(rating.text)
    else:
        return np.nan
```

```python
def get_rating_count(soup):
    rating_count = soup.find('button', class_='_1qf7wt4w')
    if rating_count != None:
        return rating_count.text.strip('()reviews ')
    else:
        return None
```

```python
def get_city_country(soup):
    condition = soup.find('span', class_ = '_pbq7fmm')
    if condition != None:
        x = condition.text.split(',')[-2:]
        return x[0].strip(), x[1].strip()

    return None, None
```

```python
def get_superhost(soup):
    condition = soup.find('span', class_='_63km3vu')
    superhost = False
    if condition != None:
        superhost = True
        
    return superhost
```

```python
def get_price(soup):
    
    s = soup.find('span', class_='_tyxjp1')
    if s != None:
        return price_to_int(s.text)
    else:
        return np.nan
```

```python
def get_dates(soup):
    start_date = soup.find('div', {'class':'_1g8031c', 'data-testid':'change-dates-checkIn'})
    end_date = soup.find('div', {'class':'_1g8031c', 'data-testid':'change-dates-checkOut'})
    if start_date != None and end_date != None:
        return str_to_date(start_date.text), str_to_date(end_date.text)
    else:
        return None, None
```

```python
def get_rooms_details(soup):
    temp = soup.find('div', {'class':'_tqmy57'})
    dictionary = {}
    if temp != None:
        temp = temp.find_all('span')
        details = list(set([x.text.strip(' ,·') for x in temp]))
        details.remove('')
        for d in details:
            t = d.split(' ')
            dictionary[t[1]] = int(t[0])
    return dictionary
```

```python
def TEST(URL):
    driver = driver_set_up_firefox()
    print('finished setting up firefox')
    
    driver.get(URL)
    print('opened the URL')
    
    scroll_page(driver)
    print('finished scrolling down page')
    
    time.sleep(3)
    L= []
    
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    print('loaded the soup')
    
    #df['rating'] = get_rating(soup)
    L.append(get_rating(soup))
    L.append(get_rating_count(soup))
    city, country = get_city_country(soup)
    L.append(city)
    L.append(country)
    L.append(get_superhost(soup))
    L.append(get_price(soup))
    start_date, end_date = get_dates(soup)
    L.append(start_date)
    L.append(end_date)
    L.append(get_rooms_details(soup))
    
    
    driver.close()
    print("finished")
    return L
```
